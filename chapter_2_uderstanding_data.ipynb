#JSON

```python
# prompt: load Json flie name data_station

import json
import pandas as pd

def load_json_file(file_path):
  """Loads a JSON file and returns its contents as a Python dictionary.

  Args:
      file_path: The path to the JSON file.

  Returns:
      A Python dictionary representing the JSON data, or None if an error occurs.
  """
  try:
    with open(file_path, 'r') as f:
      data = json.load(f)
    return data
  except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
    return None
  except json.JSONDecodeError:
    print(f"Error: Invalid JSON format in {file_path}")
    return None
  except Exception as e:
    print(f"An unexpected error occurred: {e}")
    return None

# Example usage:
file_path = '/content/drive/MyDrive/bsc_dpdm24_data/data_station.json'  # Replace with the actual file path
data = load_json_file(file_path)

if data:
  # Process the JSON data (example: convert to DataFrame)
  try:
    df_json = pd.DataFrame(data)  # Assumes JSON data represents a list of dictionaries or a dictionary of lists
    df_json.head()
  except Exception as e:
    print("Error converting JSON data to DataFrame:", e)
    print("JSON data loaded successfully. Printing raw data:")
data
```

#Distance

```python
# prompt: build pd dataframe from[[1,2],[-2,4].[0,3],[-3,-2]]

import pandas as pd

data = [[1, 2], [-2, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1', 'X2'])
df
```

```python
# prompt: draw scatter plot of the df,draw bold line X1 and X2 axes

import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with 'X1' and 'X2' columns

min_val = min(df['X1'], df['X2'])

plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(df['X1'], df['X2'],c='r')
plt.xlabel("X1", fontweight='bold')
plt.ylabel("X2", fontweight='bold')
plt.plot([-3,1],[0,0], color='black', linewidth=2)
plt.plot([0,0],[-2,4], color='black', linewidth=2)
plt.title("Scatter Plot of X1 vs X2")
plt.grid(True)
plt.show()
```

```python
# prompt: calculate distance matrix of the df

from scipy.spatial.distance import pdist, squareform

# Calculate the pairwise distances between all rows in the DataFrame
distances = pdist(df[['X1', 'X2']])

# Convert the condensed distance matrix to a square distance matrix
distance_matrix = squareform(distances)

distance_matrix
```

```python
# prompt: calculate distance matrix of the df use L1-norm

from scipy.spatial.distance import pdist, squareform

# Calculate the pairwise distances between all rows in the DataFrame using the L1 norm (Manhattan distance)
distances = pdist(df[['X1', 'X2']], metric='cityblock')

# Convert the condensed distance matrix to a square distance matrix
distance_matrix = squareform(distances)

distance_matrix
```

##Z-score

```python
# prompt: calculate z-score for df2

from scipy.stats import zscore

# Calculate z-scores for all columns in df2
df2_zscore = df2.apply(zscore)

df2_zscore
```

```python
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(df2_zscore['X1'], df2_zscore['X2'],c='r')
plt.xlabel("X1", fontweight='bold')
plt.ylabel("X2", fontweight='bold')
# plt.plot([-3,1],[0,0], color='black', linewidth=2)
# plt.plot([0,0],[-2,40], color='black', linewidth=2)
plt.title("Scatter Plot of X1 vs X2")
plt.grid(True)
plt.show()
```

```python
# prompt: do min-max normalization for the df2 for each column separately

# Min-Max normalization for each column in df2
df2_min_max = df2.copy()  # Create a copy to avoid modifying the original DataFrame

for column in df2_min_max.columns:
    min_val = df2_min_max[column].min()
    max_val = df2_min_max[column].max()
    df2_min_max[column] = (df2_min_max[column] - min_val) / (max_val - min_val)

df2_min_max
```
