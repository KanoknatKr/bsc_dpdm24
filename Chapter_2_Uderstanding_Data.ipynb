#JSON

```python
# prompt: load Json flie name data_station

import json
import pandas as pd

def load_json_file(file_path):
  """Loads a JSON file and returns its contents as a Python dictionary.

  Args:
      file_path: The path to the JSON file.

  Returns:
      A Python dictionary representing the JSON data, or None if an error occurs.
  """
  try:
    with open(file_path, 'r') as f:
      data = json.load(f)
    return data
  except FileNotFoundError:
    print(f"Error: File not found at {file_path}")
    return None
  except json.JSONDecodeError:
    print(f"Error: Invalid JSON format in {file_path}")
    return None
  except Exception as e:
    print(f"An unexpected error occurred: {e}")
    return None

# Example usage:
file_path = '/content/drive/MyDrive/bsc_dpdm24_data/data_station.json'  # Replace with the actual file path
data = load_json_file(file_path)

if data:
  # Process the JSON data (example: convert to DataFrame)
  try:
    df_json = pd.DataFrame(data)  # Assumes JSON data represents a list of dictionaries or a dictionary of lists
    df_json.head()
  except Exception as e:
    print("Error converting JSON data to DataFrame:", e)
    print("JSON data loaded successfully. Printing raw data:")
data
```
